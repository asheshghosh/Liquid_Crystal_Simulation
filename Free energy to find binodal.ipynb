```python
import matplotlib.pyplot as plt
import numpy as np
import wlcstat.wlc_lcpoly as wlc_lcpoly

lam_0 = 1e-2
lam_f = 1e2
n_lam = 500
lam_vec = np.linspace(lam_0, lam_f, n_lam)
length_kuhn = 3.3333
alpha = 10

%config InlineBackend.figure_format = 'svg'
plt.figure(figsize=(6,5))

font = {'family' : 'serif',
        'weight':'normal',
        'size': 21}
plt.rc('font', **font)

kappa_vec = np.array([20.7582,21.0606,23.6844])
for i_kap in range(len(kappa_vec)):
    kappa = kappa_vec[i_kap]
    m_val = np.zeros(n_lam)
    q_val = np.zeros(n_lam)
    f_val = np.zeros(n_lam)
    for i_lam in range(n_lam):
        lam = lam_vec[i_lam]
        q_val[i_lam] = wlc_lcpoly.q_lcpoly(length_kuhn,lam,alpha)
        m_val[i_lam] = lam / kappa
        f_val[i_lam] = kappa * length_kuhn * m_val[i_lam] ** 2 / 3 - np.log(q_val[i_lam])
    plt.plot(m_val, f_val,'-')
plt.plot(m_val, f_val*0.00,':')
plt.ylabel(r'Change in free energy, $\beta \Delta f$')
plt.xlabel(r'Order parameter, $m$')
plt.ylim((-0.6, 0.6))
plt.xlim((0.0, 0.75))
plt.tight_layout()
plt.savefig('free_energy.pdf')
plt.show()
```
